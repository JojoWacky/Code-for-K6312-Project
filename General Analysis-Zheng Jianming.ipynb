```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned.csv', index_col=0).drop(['track_id'], axis=1)
```

```python
df.groupby("year")["speechiness"].mean()
```

```python
df.groupby("year")["speechiness"].mean().plot(figsize=(10,10),title="speechiness")
```

```python
df.groupby("year")["speechiness"].mean()[20:].plot(figsize=(10,10),title="speechiness",kind="pie")
```

```python
s = df.groupby("year")["loudness"].mean()
s

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
x = list(s.index)
k1 = list(s)
plt.plot(x,k1,'s-',color = 'r',label="line")
plt.bar(x,k1,color = 'g',label="bar")
plt.xlabel("year")
plt.ylabel("valence")
plt.legend(loc = "best")
plt.show()
```
