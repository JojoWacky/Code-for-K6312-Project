```python
from os import path
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from wordcloud import WordCloud, ImageColorGenerator
```

```python
with open('style.txt') as fp:
    text = fp.read()
    mask = np.array(Image.open('MARK.jpeg'))
    wordcloud = WordCloud(background_color="white",max_words = 2000,mask = mask).generate(text)
    image_produce= wordcloud.to_image()
    image_produce.show()
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.figure()
    plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
df = pd.read_csv("bad guys.csv")
```

```python
labels=np.array(["Speechiness","Danceability","Energy","Acousticness","Liveness"])
stats=df.loc[0,labels].values
labs = ["Speechiness","Danceability","Energy","Acousticness","Liveness"]
```

```python
# Make some calculations for the plot

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

# Plot stuff
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats,'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
ax.set_thetagrids(angles*180/np.pi, stats, fontproperties="Microsoft Yahei")
ax.set_rlabel_position (15)

plt.show()
```
